```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
print("Keys of iris_dataset:\n ", iris_dataset.keys())
```

```python
iris_dataset
```

```python
print("Target names:",iris_dataset['target_names'])
```

```python
print("feature names:",iris_dataset['feature_names'])
```

```python
print("Target:",iris_dataset['target'])
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], test_size = 0.20 , random_state=0)
```

```python
print("X_train shape:", X_train.shape)
print("Y_train shape:", y_train.shape)
```

```python
print("X_test_shape :" , X_test.shape)
print("y_test_shape :" , y_test.shape)
```

```python
import pandas as pd
import sys
!pip install mglearn
import mglearn as mg
iris_dataframe = pd.DataFrame(X_train , columns = iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train , figsize=(15,15), marker = 'o' , hist_kwds={'bins':10}, s=60 , alpha = 0.8 , cmap = mg.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

Prediction

```python
import numpy as np
X_new = np.array([[5 , 2.9 , 1 , 0.2]])
print("X_new.shape:",X_new.shape)
```

```python
prediction = knn.predict(X_new)
print("Prediction:", prediction)
print("Predicted target name :", iris_dataset['target_names'][prediction])
```

eval

```python
y_pred = knn.predict(X_test)
print("Test set prediction", y_pred)
```

```python
print("test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print("Accuracy score ")
print(accuracy_score(y_test , y_pred))
print("\n")
print("classification report \n")
print(classification_report(y_test , y_pred))
print("confusion matrix \n")
print(confusion_matrix(y_test , y_pred))
```
